```python
#Relacion de peso y dmg
# 1. SelecciÃ³n y limpieza de columnas numÃ©ricas
damage_cols = ['weight', 'attack_Ligt_amount', 'attack_Phy_amount', 'attack_Mag_amount', 'attack_Fire_amount', 'attack_Holy_amount', 'attack_Crit_amount']
df_numeric = df_weapons2[damage_cols].apply(pd.to_numeric, errors='coerce').dropna()

# 2. Matriz de correlaciÃ³n
corr_matrix = df_numeric.corr()

# 3. Heatmap mejorado
plt.figure(figsize=(10, 8))
sns.heatmap(
    data=corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    vmin=-1,  # Rango fijo para mejor interpretaciÃ³n
    vmax=1,
    linewidths=0.5,
)
plt.title("CorrelaciÃ³n entre Peso y Tipos de DaÃ±o", pad=20)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
