## Conteo de Bosses por region

```python
def parse_bosses(x):
    if isinstance(x, float) and np.isnan(x):
        return []
    elif isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return [] 
# Agrupar por región y contar jefes
df_combined['boss_count'] = df_combined['bosses'].apply(lambda x: len(x) if isinstance(x, list) else 0)
bosses_per_region = df_combined.groupby('region')['boss_count'].sum()

# Prepara los datos
labels = [f"{region} ({count})" for region, count in bosses_per_region.items()]
sizes = bosses_per_region.values
colors = plt.cm.tab20.colors  # Puedes usar otro colormap como 'Set3', 'Accent', etc.

# Grafico
fig, ax = plt.subplots(figsize=(10, 10))
wedges, texts = ax.pie(sizes, colors=colors[:len(sizes)], startangle=140)

# Leyendas
ax.legend(wedges, labels,
          title="Regiones",
          loc="center left",
          bbox_to_anchor=(1, 0.5))

ax.set_title("Conteo de Bosses por Región")
plt.tight_layout()
plt.show()
```
