## Debilidades de Bosses Principales de Elden Ring


```python
df_bosseswr=dataframes['bosses_final_clean']
```

```python
df_exploded = df_bosseswr.explode('weakness')
```

```python
weakness_counts = df_exploded['weakness'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=weakness_counts.index, y=weakness_counts.values, palette="magma")
plt.title('Cantidad de jefes débiles por tipo de ataque')
plt.xlabel('Tipo de debilidad')
plt.ylabel('Número de jefes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Tabla dinámica: filas = debilidad, columnas = jefes (1 si aplica, NaN si no)
pivot = df_exploded.assign(value=1).pivot_table(index='weakness', columns='name', values='value', fill_value=0)

# Heatmap para mostrar asociación
plt.figure(figsize=(10, 6))
sns.heatmap(pivot, annot=True, cmap="YlGnBu", cbar=False, linewidths=0.5)
plt.title("Jefes por tipo de debilidad")
plt.ylabel("Tipo de debilidad")
plt.xlabel("Jefe")
plt.tight_layout()
plt.show()
```

Se puede observar que hay mas jefes que son debiles a ataques fisicos que con efectos.
